```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import logging
import time
import wandb
wandb.login()
import random
```

```python
model_names=['b20_e100_r001_adam_mse_nnconvMax_poolMeanAggr_2fc','b20_e100_r001_adam_mse_nnconvMean_poolMaxAggr_2fc','b20_e100_r001_sgd_mse_nnconvMean_poolMeanAggr_2fc', 'b20_e100_r005_adam_mse_nnconvMean_poolMeanAggr_2fc', 'b50_e100_r001_adam_mse_nnconvMean_poolMaxAggr_2fc_wandb','b20_e100_r001_adam_mse_nnconv_poolMeanAggr_2fc','b128_e100_r001_adam_mse','b128_e135_r001_adam_mse_msgMean_poolMean_2fc','b128_e135_r001_adam_mse_set2set_2fc', 'b20_e50_r001_adam_mse_nnconvAdd_poolMeanAggr_3fc']
colors=['y','b','r','purple','g','orange','grey','w','brown','lime']
```

```python
num_models=len(model_names)
loss_values=[None]*num_models
val_loss_values=[None]*num_models
```

```python
for index,model_name in enumerate(model_names):
    loss_values[index] = np.loadtxt("./loss_values/loss_values_"+model_name+".txt")
    val_loss_values[index] = np.loadtxt("./loss_values/val_loss_values_"+model_name+".txt")
```

```python
plt.figure(figsize=(8, 6)) 
for index,model_name in enumerate(model_names):
    plt.plot(loss_values[index], colors[index],label=model_names[index])
plt.xlabel('epoch') 
plt.ylabel('loss') 
plt.title("Train Loss")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6)) 
for index,model_name in enumerate(model_names):
    plt.plot(val_loss_values[index], colors[index],label=model_names[index])
plt.xlabel('epoch') 
plt.ylabel('loss') 
plt.title("Validation Loss")
plt.legend()
plt.show()
```

```python
# Initialize wandb
model_name = 'b20_e100_r001_adam_mse_nnconvMax_poolMeanAggr_2fc'
wandb.init(project='combination', entity=model_name)

# Read loss values from text file
with open('loss_values_'+model_name+''.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        loss = float(line.strip())
        wandb.log({'loss': loss})

# Finish the run
wandb.finish()
```
